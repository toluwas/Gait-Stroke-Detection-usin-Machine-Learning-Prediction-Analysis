```python
import pandas as pd 
import numpy as np
gait_dataset = pd.read_csv('gait_features_all_subjects_with_stroke.csv')
gait_dataset
```

```python
gait_dataset = gait_dataset.drop(['subject', 'speed'], axis=1)
gait_dataset.head()
```

```python
# Artificial Neural Networks with LDA
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, roc_auc_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import scikitplot as skplt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Import the dataset
gait_dataset

# Separate features and target
X = gait_dataset.drop('target', axis=1)  # Independent variables
y = gait_dataset['target']               # Dependent variable

# Feature Scaling
# Standardize the features
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

# Apply LDA (Supervised Dimensionality Reduction)
lda = LinearDiscriminantAnalysis(n_components=1)  # Only 1 component because it's binary classification
X_lda = lda.fit_transform(X_scaled, y)

# Train-test split using LDA-reduced features
X_trainval, X_test, y_trainval, y_test = train_test_split(X_lda, y, test_size=0.2, stratify=y, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.25, stratify=y_trainval, random_state=0)

# Build the ANN
classifier = Sequential()
classifier.add(Dense(units=9, activation='relu', input_dim=1))  # Input layer (1 LDA component)
classifier.add(Dense(units=9, activation='relu'))               # Hidden layer
classifier.add(Dense(units=1, activation='sigmoid'))            # Output layer

# Compile
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Fit the model
history = classifier.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=10)

# Predicting the Test set results
y_pred_proba = classifier.predict(X_test)  # Probabilities
y_pred = (y_pred_proba > 0.5)              # Binary predictions

# Confusion Matrix and Report
cm = confusion_matrix(y_test, y_pred)
print("\nANN Classifier (LDA):")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(cm)

# Per-class Accuracy
per_class_accuracies = cm.diagonal() / cm.sum(axis=1)
for cls, acc in enumerate(per_class_accuracies):
    print(f"Accuracy for class {cls}: {acc:.4f}")

# Confusion Matrix Visualization
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=False, title='Confusion Matrix for ANN Classifier with LDA')
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (LDA Features)')
plt.legend(loc='lower right')
plt.show()

# Final AUC Score
auc_value = roc_auc_score(y_test, y_pred_proba)
print(f"AUC: {auc_value:.2f}")
```

```python
# SVC Classifier with LDA
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import scikitplot as skplt
import matplotlib.pyplot as plt

# Import the dataset
gait_dataset

# Separate features and target
X = gait_dataset.drop('target', axis=1)  # Independent variables
y = gait_dataset['target']               # Dependent variable

# 1. Split dataset
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.25, random_state=0, stratify=y_trainval)

# 2. Standardize features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_val = sc.transform(X_val)
X_test = sc.transform(X_test)

# 3. Apply LDA (n_components=1 for binary classification)
lda = LinearDiscriminantAnalysis(n_components=1)
X_train_lda = lda.fit_transform(X_train, y_train)
X_val_lda = lda.transform(X_val)
X_test_lda = lda.transform(X_test)

# 4. Train LinearSVC with probability calibration
svc = LinearSVC(C=5, random_state=0, max_iter=10000)
svc_model = CalibratedClassifierCV(svc, method='sigmoid')
svc_model.fit(X_train_lda, y_train)

# 5. Cross-validation
scores = cross_val_score(svc_model, lda.transform(sc.transform(X_trainval)), y_trainval, cv=5)
print("Cross Validation Scores:", scores)
print("Average CV Score:", scores.mean())

# 6. Validation prediction
y_pred_val = svc_model.predict(X_val_lda)
print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))

# 7. Test prediction
y_pred_test = svc_model.predict(X_test_lda)
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))

# 8. Evaluation
print("\nSVC Classifier with LDA:")
print(classification_report(y_test, y_pred_test))
cm = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix:")
print(cm)
per_class_accuracies = cm.diagonal() / cm.sum(axis=1)
for cls, acc in enumerate(per_class_accuracies):
    print(f"Accuracy for class {cls}: {acc:.4f}")

# 9. Confusion Matrix Visualization
skplt.metrics.plot_confusion_matrix(y_test, y_pred_test, normalize=False, title='Confusion Matrix for SVC (LDA)')
plt.show()

# 10. ROC Curve
y_pred_proba = svc_model.predict_proba(X_test_lda)
skplt.metrics.plot_roc(y_test, y_pred_proba, title='ROC Curve for SVC (LDA)')
plt.show()

# 11. AUC
auc_value = roc_auc_score(y_test, y_pred_proba[:, 1])
print(f"AUC SVC (LDA): {auc_value:.2f}")
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import scikitplot as skplt

# Import the dataset
gait_dataset

# Separate features and target
X = gait_dataset.drop('target', axis=1)  # Independent variables
y = gait_dataset['target']               # Dependent variable

# STEP 1: Split original dataset
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.25, stratify=y_trainval, random_state=0)

# STEP 2: Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# STEP 3: Apply LDA (1 component for binary classification)
lda = LinearDiscriminantAnalysis(n_components=1)
X_train_lda = lda.fit_transform(X_train_scaled, y_train)
X_val_lda = lda.transform(X_val_scaled)
X_test_lda = lda.transform(X_test_scaled)

# STEP 4: Train Random Forest on LDA features
rf = RandomForestClassifier(n_estimators=100, random_state=0)
rf.fit(X_train_lda, y_train)

# STEP 5: Make predictions
y_pred_val = rf.predict(X_val_lda)
y_pred_test = rf.predict(X_test_lda)
y_pred_proba = rf.predict_proba(X_test_lda)[:, 1]

# STEP 6: Evaluate performance
print("Validation Accuracy:", rf.score(X_val_lda, y_val))
print("Test Accuracy:", rf.score(X_test_lda, y_test))

print("\nClassification Report (Test Set):")
print(classification_report(y_test, y_pred_test))

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred_test)
print(cm)

# STEP 7: Visualizations
# Confusion Matrix
skplt.metrics.plot_confusion_matrix(y_test, y_pred_test, normalize=False, title='Confusion Matrix for RF with LDA')
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = roc_auc_score(y_test, y_pred_proba)

plt.figure()
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='green')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Random Forest (LDA Features)')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"AUC Score: {roc_auc:.2f}")
```
